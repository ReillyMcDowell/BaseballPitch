```python
import os

input_dir = "D:/GitHub/BaseballPitch/modules/pitcher_segmentation/finetuning_dataset/pose_labels/val"
output_dir = "D:/GitHub/BaseballPitch/modules/pitcher_segmentation/finetuning_dataset/detect_labels/val"

file_count = 0
for root, _, files in os.walk(input_dir):
    for file in files:
        if file.endswith(".txt"):
            in_path = os.path.join(root, file)
            rel_path = os.path.relpath(in_path, input_dir)
            out_path = os.path.join(output_dir, rel_path)
            os.makedirs(os.path.dirname(out_path), exist_ok=True)
            with open(in_path, "r") as fin, open(out_path, "w") as fout:
                for line in fin:
                    parts = line.strip().split()
                    if len(parts) >= 5:
                        fout.write(" ".join(parts[:5]) + "\n")
            print(f"Processed: {rel_path}")
            file_count += 1

if file_count == 0:
    print("No .txt label files found in input directory.")
else:
    print(f"Done! {file_count} label files processed.")
```

```python
import torch
from ultralytics import YOLO

# Load a model
model = YOLO("yolo26n.yaml")  # build a new model from YAML
model = YOLO("yolo26n.pt")  # load a pretrained model (recommended for training)
model = YOLO("yolo26n.yaml").load("yolo26n.pt")  # build from YAML and transfer weights

# Check GPU availability
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Training on device: {device}")
if device == 'cpu':
    print("⚠️  GPU not available. Training will be slow on CPU.")

# Train the model
results = model.train(data="D:/GitHub/BaseballPitch/modules/pitcher_segmentation/runs/pitcher_detect.yaml", 
                      epochs=1, imgsz=1280, plots=True, name='pitcher_pose_yolo26n',
                      project='D:/GitHub/BaseballPitch/modules/pitcher_segmentation/runs/detect')
```
