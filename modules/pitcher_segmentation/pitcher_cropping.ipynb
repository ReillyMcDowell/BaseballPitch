```python
from pathlib import Path
import os
import cv2

from ultralytics import YOLO, solutions

video_path = Path("D:/GitHub/BaseballPitch/test_pitch_video.mp4").resolve()
print("cwd:", os.getcwd())
print("video_path:", video_path)
print("exists:", video_path.exists())

if not video_path.exists():
    raise FileNotFoundError(f"Video not found: {video_path}")

cap = cv2.VideoCapture(str(video_path))
if not cap.isOpened():
    raise RuntimeError(
        "OpenCV cannot open this video. Check path/filename."
    )

# Initialize object cropper
cropper = solutions.ObjectCropper(
    show=False,  # display the output
    model="D:/GitHub/BaseballPitch/yolo26n.pt",  # model for object cropping, e.g., yolo26x.pt.
    classes=[0],  # crop specific classes such as person (label 0) COCO pretrained model.
    conf=0.5,  # adjust confidence threshold for the objects.
    crop_dir="D:/GitHub/BaseballPitch/test_output",  # set the directory name for cropped detections
)
model=YOLO("D:/GitHub/BaseballPitch/yolo26n.pt")  # model for object cropping, e.g., yolo26x.pt.
# Set the desired frame width, height, and FPS
frame_width = 640  # Set your desired width
frame_height = 480  # Set your desired height
fps = 60  # Set your desired FPS

# Create a VideoWriter object to save the modified video
video_writer = cv2.VideoWriter("modified_video.avi", cv2.VideoWriter_fourcc(*"mp4v"), fps, (frame_width, frame_height))

# Process video
while cap.isOpened():
    success, im0 = cap.read()

    if not success:
        print("Video frame is empty or processing is complete.")
        break
    
    results = model.predict(im0)
    
    # Process the results as needed
    for result in results:
        boxes = result.boxes.xyxy.cpu().tolist()
    
    cropped_frame = cropper.process(im0)
    print(f"Frame type: {type(cropped_frame.plot_im)}, shape: {cropped_frame.plot_im.shape}")
    cv2.imshow("Cropped Video", cropped_frame.plot_im)
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break
    # Resize the frame to the desired dimensions
    resized_frame = cv2.resize(cropped_frame.plot_im, (frame_width, frame_height))
    print(f"Frame type: {type(resized_frame)}, shape: {resized_frame.shape}")
    
    # Write the resized frame to the output video
    video_writer.write(resized_frame)

    # Display the frame (optional)
    # cv2.imshow("Cropped Video", resized_frame)
    # if cv2.waitKey(1) & 0xFF == ord("q"):
    #     break

cap.release()
video_writer.release()
cv2.destroyAllWindows()  # destroy all opened windows
```

```python
import cv2
from pathlib import Path
import os
from ultralytics import YOLO
model = YOLO("D:/GitHub/BaseballPitch/yolo26n-pose.pt")

# Open the video file
video_path = Path("D:/GitHub/BaseballPitch/test_pitch_video.mp4").resolve()
print("cwd:", os.getcwd())
print("video_path:", video_path)
print("exists:", video_path.exists())

if not video_path.exists():
    raise FileNotFoundError(f"Video not found: {video_path}")

cap = cv2.VideoCapture(str(video_path))
if not cap.isOpened():
    raise RuntimeError(
        "OpenCV cannot open this video. Check path/filename."
    )

# Loop through the video frames
while cap.isOpened():
    # Read a frame from the video
    success, frame = cap.read()

    if success:
        # Run YOLO26 tracking on the frame, persisting tracks between frames
        results = model.track(frame, persist=True)

        # Visualize the results on the frame
        annotated_frame = results[0].plot()

        # Display the annotated frame
        cv2.imshow("YOLO26 Tracking", annotated_frame)

        # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        # Break the loop if the end of the video is reached
        break

# Release the video capture object and close the display window
cap.release()
cv2.destroyAllWindows()
```
